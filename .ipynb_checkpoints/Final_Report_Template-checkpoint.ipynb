## Introduction

<!-- Include your motivation and any background information needed to understand your research question and data analysis. Clearly state your research question and hypothesis. -->

As humans continue to grow as a species, we have been burning more and more fossil fuels as means of energy production. As well as an increase of global temperature, CO2 dissolves into the ocean and causes shifts in marine chemistry. This can have important effects on the health of marine ecosystems and the people that rely on them. In addition to economic and human consequences, coral reefs, the ecosystems most threatened by these chemical changes, hold a massive amount of the ocean’s known biodiversity. This analysis of data seeks to try to quantify to what extent anthropogenic CO2 inputs have affected dissolved inorganic carbon in the ocean. Given the changes we have seen in the oceans in the past 50 years, my hypothesis is that increased CO2 will cause an increase in marine DIC.

The figures generated by the marine CO2 data show a strong positive trend in DIC in the past 40 years.
